```python
date_init = pd.date_range(start="24/08/2021", end="24/10/2021", freq="D")
date_init
```

```python
date_end = pd.date_range(start="24/11/2021", end="24/01/2022", freq="D")
date_end
```

```python
list1 = []
for i in range(15):
    list1.append(np.random.randint(2))
df = pd.DataFrame()
df["Campaña_inicio"] = date_init[2:17]
df["Campaña_fin"] = date_end[2:17]
df["Frecuencia"] = list1
df
```

```python
df["Día de inicio"] = df["Campaña_inicio"].dt.day
df["Mes de inicio"] = df["Campaña_inicio"].dt.month
df["Año de inicio"] = df["Campaña_inicio"].dt.year
df["Hora de inicio"] = df["Campaña_inicio"].dt.second
df["Minuto de inicio"] = df["Campaña_inicio"].dt.minute
df["Segundo de inicio"] = df["Campaña_inicio"].dt.second
df["Día de la semana de inicio"] = df["Campaña_inicio"].dt.weekday

df["Duración de campaña"] = df["Campaña_fin"] - df["Campaña_inicio"]
df
```

```python
from datetime import timedelta
df["Días de duración"] = df["Duración de campaña"] / timedelta(days=1)
df["Minutos de duración"] = df["Duración de campaña"] / timedelta(minutes=1)
df["Segundos de duración"] = df["Duración de campaña"] / timedelta(seconds=1)

df
```
