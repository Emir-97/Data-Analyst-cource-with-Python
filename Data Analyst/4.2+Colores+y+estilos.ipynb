```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
fig, pan_1 = plt.subplots(1, 1)

x1 = np.linspace(0, 4 * np.pi, 100)
y1 = np.sin(x1)

#Con linestyle="" se establece el tipo de lineado que tendrá la grafica y con color="" el color de la misma y 
#solo se coloca la inicial en ingles
pan_1.plot(x1, y1, linestyle = '--', color = 'g')
```

```python
fig, pan_1 = plt.subplots(1, 1)

x1 = np.linspace(0, 4 * np.pi, 100)
y1 = np.sin(x1)

#Con marker="" se establece la forma que tendrá los puntos que superpongan la linea
#Con markeredgecolor="" se establece el color del contorno o borde de las marcas
#Con markerfacecolor="" fija el color del centro de las marcas
pan_1.plot(x1, y1, linestyle = '--', color = 'y', marker = 'o', markeredgecolor = 'y', markerfacecolor = 'r')
```
